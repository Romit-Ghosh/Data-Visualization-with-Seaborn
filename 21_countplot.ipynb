# Seaborn: countplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
diamonds = sns.load_dataset('diamonds')

diamonds.head()
```

```python
diamonds = diamonds[diamonds.clarity.isin(['SI1', 'VS2'])]

diamonds
```

```python
diamonds.shape
```

## Basics

```python
sns.set_style('darkgrid')
```

```python
sns.countplot(x='color', data=diamonds);
```

```python
diamonds.color.value_counts(sort=False)
```

```python
diamonds.color.value_counts(sort= True)
```

### Horizontal Countplot

```python
sns.countplot(y='color', data=diamonds);
```

## Options

### `order`

```python
sns.countplot(x='color', data=diamonds);
```

```python
sns.countplot(x='cut', data=diamonds);
```

```python
sns.countplot(x='cut', data=diamonds);

plt.tick_params(
    labelbottom= False,
    labeltop= True
)
```

```python
color_order=['J', 'I', 'H', 'G', 'F', 'E', 'D']  # for customized ordering of the plot

sns.countplot(x='color', data=diamonds, order=color_order);
```

### Order Ascending or Descending

```python
diamonds.color.value_counts().index
```

```python
sns.countplot(x='color', data=diamonds,
              order=diamonds.color.value_counts().index  # ascending
             ); 
```

```python
sns.countplot(x='color', data=diamonds,
              order=diamonds.color.value_counts().index[::-1]  #remove [::-1] for descending
             ); 
```

### `hue`, `hue_order`

```python
sns.countplot(x='color', data=diamonds, hue='clarity', hue_order=['SI1', 'VS2']);
```

## Styling

### `color` and `palette`

```python
sns.countplot(x='color', data=diamonds, color='skyblue');  # color parameter so as to make all the bars look the same color
```

```python
sns.countplot(x='color', data=diamonds, palette='twilight');
```

```python
diamonds.color.value_counts()
```

### Other Matplotlib Properties

```python
sns.countplot(x='color', data=diamonds,
              lw=6, ec='black', hatch='/'  # ec= edge color
             );  
```

## Examples, Thumbnail

```python
from matplotlib import pyplot as plt
```

```python
sns.set(style='darkgrid', font_scale=1.3)
sns.countplot(x='cut', data=diamonds, color='blue', lw=3, ec= 'yellow')
plt.xlabel('')
plt.ylabel('Count')


plt.tight_layout()
#plt.savefig('20countplot_image_countplotExample.png', dpi=200);
```

```python
sns.set_style('white')

import pandas as pd

df = pd.DataFrame({'Animal': ['dog', 'cat', 'cat', 'dog', 'dog', 'cat', 'dog', 'cow'],
                   'Weight': [55, 8, 12, 25, 100, 10, 60, 260]
                  })
```

```python
df
```

```python
plt.figure(figsize=(3, 3))

sns.countplot(x='Animal', data=df, ec='black', lw=3, palette=['darkred', '#2F8F8F', 'blue'])

plt.yticks([0, 1, 2, 3, 4])
plt.ylabel('Count', labelpad=20)

sns.despine()

plt.tight_layout()
#plt.savefig('21countplot_figure_dogsCats.svg');
```

```python
result = df.groupby('Animal').Weight.mean().to_frame().rename_axis('A')
result
```

```python
result = df.groupby('Animal').Weight.mean().to_frame().reset_index().rename(columns={'Animal': 'A'})
result
```

```python
plt.figure(figsize=(5, 6))

sns.barplot(x='Animal', y='Weight', data=df, ec='black', lw=3, 
            palette=['darkred', '#2F8F8F'],
            errorbar=None
           )

plt.ylim(0, 75)
plt.ylabel('Weight', labelpad=20)

sns.despine()

plt.tight_layout()
#plt.savefig('21countplot_figure_dogsCats_bar_noCI.svg');
```

```python
plt.figure(figsize=(3, 6))

sns.barplot(x='Animal', y='Weight', data=df, ec='black', lw=3, 
            palette=['darkred', '#2F8F8F'],
            errorbar=('ci', 68), errcolor='black'
           )

plt.ylim(0, 75)
plt.ylabel('Weight', labelpad=20)

sns.despine()

plt.tight_layout()
#plt.savefig('21countplot_figure_dogsCats_bar_withCI.svg');
```

### Thumb

```python
plt.figure(figsize=(7,4))
sns.set(style='white', font_scale=1.2)

ax = sns.countplot(
                   x='color', data=diamonds, 
                   hue='clarity', hue_order=['SI1', 'VS2'], 
                   palette=['#2F8F8F', 'lightgray']
                  )

ax.legend([], frameon=False)
sns.despine()

plt.xlabel('')
plt.ylabel('Count', labelpad=10)
plt.tight_layout()
plt.savefig('21countplot_image_thumb.png', dpi=260);
```
